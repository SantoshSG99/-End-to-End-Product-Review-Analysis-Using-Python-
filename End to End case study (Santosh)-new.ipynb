```python
# Calculate customer retention month-on-month
# We need to find out if customers who purchased in one month return in subsequent months

# Create a dataframe that marks each customer's purchase by month and year
monthly_purchases = f_m.groupby(['customer_unique_id', 'purchace_month']).size().reset_index(name='purchases')

# Pivot this data to have customers as rows and months as columns
pivot_purchases = monthly_purchases.pivot_table(index='customer_unique_id', columns=['purchace_month'], values='purchases', fill_value=0)

# Calculate retention by checking if customers who purchased in one month also purchased in any subsequent months
retention_matrix = pivot_purchases.apply(lambda x: x > 0).astype(int).diff(axis=1).fillna(0).applymap(lambda x: 1 if x > 0 else 0)

retention_matrix= retention_matrix.reset_index()
retention_matrix
```

### '0' indicates no purchase in the subsequent month

## 3. Cross-Selling (Which products are selling together)
### Hint: We need to find which of the top 10 combinations of products are selling together in  each transaction. (combination of 2 or 3 buying togethe

```python
grouped_data = f_m.groupby('order_id')['product_category_name'].apply(list)
def generate_combinations(product_list, size):
    combinations = []
    length = len(product_list)
    for i in range(length):
        for j in range(i + 1, length):
            if size == 2:
                if product_list[i] != product_list[j]:
                    combinations.append((product_list[i], product_list[j]))
            elif size == 3:
                for k in range(j + 1, length):
                    if product_list[i] != product_list[j] != product_list[k]:
                        combinations.append((product_list[i], product_list[j], product_list[k]))
    return combinations


combinations_2 = grouped_data.apply(lambda products: generate_combinations(products, 2))
```

```python
all_combinations_2 = [combo for sublist in combinations_2 for combo in sublist]
combination_counts_2 = pd.Series(all_combinations_2).value_counts()
```
